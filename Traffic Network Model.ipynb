A timetable and a running chart are shown below.

The timetable depicts a network of schedules spanning between 3 stations.
The numbers in row i and column j of the timetable $Hk$ imply that at $i$ o'clock, a train takes $Hk[i, j]$ hours to travel from station $Hk$ to station $Hj$.
If $Hk[i, j]$ corresponds to $\infty$, then no train travels the corresponding route at that time.  

For example, at 8:00, a train departs from station $H0$ to $H1$, which takes 2.42 hours.

In the graph, nodes correspond to given stations and edges contain information about start and arrival times so that a shortest path can be searched in that graph using Dijkstra's algorithm (with slight modifications).
Instead of having simple weights, as is usually the case, edges are labeled with labels of the form $s-t$, where $s$ is the start time and $t$ is the arrival time. 

For example, the edge from station 0 to station 1 is labeled 8-10.42. It means that station 0 to 1, which departs at 8:00, arrives at 10:42.

If there are trains at different times, there can also be more than one side between the two stations. **For next day arrivals, no side is shown**.
